- task : create 3 buttons which print area of circle,rectangle and triangle 

```python
from tkinter import *
root=Tk()
root.geometry("600x400")

def circle():
    pi=3.14
    r=7
    print("Area of circle : " ,pi*r*r)

def rect():
    len=23
    wid = 12
    print("Area of rectangle : ", len*wid)

def tri():
    base=7
    height = 4
    print("Area of triangle : ", (height*base)/2)
f1=Frame(root,bg="aqua",borderwidth=6)
f1.pack(side=LEFT,anchor="nw")
f2=Frame(root,bg="grey",borderwidth=6)
f2.pack(side=RIGHT,anchor="ne")
f3=Frame(root,bg="brown",borderwidth=6)
f3.pack(side=TOP,anchor="sw")

b1=Button(f1,fg="red",text="area of circle", command=circle)
b1.pack(side=LEFT,padx=23)
b2=Button(f2,fg="red",text="area of rectangle", command=rect)
b2.pack(side=LEFT,padx=23)
b3=Button(f3,fg="red",text="area of triangle", command=tri)
b3.pack(side=LEFT,padx=45)

root.mainloop()
```

- task2 : buttons in all directions

```python
from tkinter import *
root=Tk()
root.geometry("600x400")

def north():
    print("Jammu Kashmir")

def south():
    print("Karnataka")

def west():
    print("Gujarat")
def east():
    print("Nagaland")

f1=Frame(root, bg="red", borderwidth=10)
f1.pack(side=TOP,anchor="n",fill=X)
f2=Frame(root, bg="blue", borderwidth=10)
f2.pack(side=BOTTOM,anchor="s",fill=X)
f3=Frame(root, bg="Yellow", borderwidth=10)
f3.pack(side=LEFT,anchor="nw",fill=Y)
f4=Frame(root, bg="green", borderwidth=10)
f4.pack(side=RIGHT,anchor="ne",fill=Y)

b1=Button(f1,text="North",command=north)
b1.pack(padx=23,side=

TOP)
b2=Button(f2,text="South",command=south)
b2.pack(padx=23,side=BOTTOM)
b3=Button(f3,text="West",command=west)
b3.pack(padx=23,side=LEFT)
b4=Button(f4,text="East",command=east)
b4.pack(padx=23,side=RIGHT)
root.mainloop()
```
