```python
#login form with Grid method

from tkinter import *
root= Tk()

root.geometry("400x200")
root.minsize (200,100)
root.maxsize(1200,900)

#lables
l1= Label(root,text="Username")
l2= Label(root,text="Password")

#to set labels into grid layout
l1.grid(row=0)
l2.grid(row=1)

#to set the entry for the labels :
e1=Entry(root)
e2=Entry(root)

#to set entry into grid layout
e1.grid(row=0,column=1)
e2.grid(row=1, column=1)

b1=Button(root,text="SUBMIT")
b1.grid(row=3, column=1)

root.mainloop()
```
